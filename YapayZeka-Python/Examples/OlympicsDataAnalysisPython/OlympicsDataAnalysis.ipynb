```python
# Eksik deerleri kontrol et
print(athletes.isnull().sum())

# Eksik deerleri doldurmak veya temizlemek
data = athletes.dropna()  # Eksik deerleri ieren satrlar kaldr
```

```python
# statistiksel zet
print(data.describe())

# Ya dalm
print(data['Age'].value_counts())

# Medal stunundaki farkl deerleri kontrol et
print(data['Medal'].value_counts())
```

```python
# Christine Jacoba Aaftink verilerini filtrele
aftink_data = data[data['Name'] == 'Christine Jacoba Aaftink']
print(aftink_data)
```

```python
import seaborn as sns

# Medallar saysal olarak kodlayalm: NA -> 0, 'Gold' -> 1, 'Silver' -> 2, 'Bronze' -> 3
data['Medal_Code'] = data['Medal'].map({'NA': 0, 'Gold': 1, 'Silver': 2, 'Bronze': 3})

# Ya ve madalya kodu arasndaki ilikiyi grselletirelim
plt.figure(figsize=(12, 6))
sns.boxplot(x='Medal_Code', y='Age', data=data)
plt.title('Ya ve Madalya Arasndaki liki')
plt.xlabel('Madalya Kodu')
plt.ylabel('Ya')
plt.xticks(ticks=[0, 1, 2, 3], labels=['NA', 'Gold', 'Silver', 'Bronze'])
plt.show()
```

```python
# Spor dallarna gre madalya saysn hesaplayalm
sports_medals = data.groupby('Sport')['Medal'].count().sort_values(ascending=False)

plt.figure(figsize=(12, 8))
sports_medals.plot(kind='bar')
plt.title('Spor Dallarna Gre Madalya Says')
plt.xlabel('Spor Dali')
plt.ylabel('Madalya Says')
plt.xticks(rotation=90)
plt.show()
```

```python
# lkelere gre madalya saysn hesaplayalm
country_medals = data.groupby('NOC')['Medal'].count().sort_values(ascending=False)

plt.figure(figsize=(12, 8))
country_medals.plot(kind='bar')
plt.title('lkelere Gre Madalya Says')
plt.xlabel('lke Kodu')
plt.ylabel('Madalya Says')
plt.xticks(rotation=90)
plt.show()
```

```python
# Ya gruplarna ayralm
bins = [0, 20, 30, 40, 50, 60]
labels = ['<20', '20-30', '30-40', '40-50', '50+']
data['Age_Group'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)

# Ya grubuna gre madalya kazanma orann hesaplayalm
age_group_medal_ratio = data.groupby('Age_Group')['Medal'].apply(lambda x: (x != 'NA').sum() / len(x))

plt.figure(figsize=(10, 6))
age_group_medal_ratio.plot(kind='bar')
plt.title('Ya Gruplarna Gre Madalya Kazanma Oran')
plt.xlabel('Ya Grubu')
plt.ylabel('Madalya Kazanma Oran')
plt.show()
```

```python
# Medal trne gre ortalama ya
mean_age_by_medal = data.groupby('Medal')['Age'].mean()

plt.figure(figsize=(10, 6))
mean_age_by_medal.plot(kind='bar')
plt.title('Medal Trlerine Gre Ortalama Ya')
plt.xlabel('Madalya Tr')
plt.ylabel('Ortalama Ya')
plt.xticks(rotation=0)
plt.show()
```

```python
# En fazla madalya kazanan sporcular
top_performers = data['Name'].value_counts().head(10)

plt.figure(figsize=(12, 8))
top_performers.plot(kind='bar')
plt.title('En Fazla Madalya Kazanan Sporcular')
plt.xlabel('Sporcu smi')
plt.ylabel('Madalya Says')
plt.xticks(rotation=90)
plt.show()
```

```python
# Spor dallar ve yllar arasndaki iliki
plt.figure(figsize=(14, 8))
sns.countplot(x='Year', hue='Sport', data=data, palette='tab20')
plt.title('Yllara Gre Spor Dallar')
plt.xlabel('Yl')
plt.ylabel('Say')
plt.xticks(rotation=90)
plt.show()
```

```python
# Medalyaya gre ya, boy ve kiloyu grselletirme
fig, axs = plt.subplots(3, 1, figsize=(12, 18))

# Ya
sns.boxplot(x='Medal', y='Age', data=data, ax=axs[0])
axs[0].set_title('Madalya Trne Gre Ya Dalm')

# Boy
sns.boxplot(x='Medal', y='Height', data=data, ax=axs[1])
axs[1].set_title('Madalya Trne Gre Boy Dalm')

# Kilo
sns.boxplot(x='Medal', y='Weight', data=data, ax=axs[2])
axs[2].set_title('Madalya Trne Gre Kilo Dalm')

plt.tight_layout()
plt.show()
```

```python
# Yllara gre madalya saysn hesaplayalm
medal_trends = data.groupby('Year')['Medal'].count()

plt.figure(figsize=(12, 6))
medal_trends.plot(kind='line')
plt.title('Yllara Gre Madalya Kazanm Trendleri')
plt.xlabel('Yl')
plt.ylabel('Madalya Says')
plt.grid(True)
plt.show()
```

```python
# lkelere gre madalya trlerinin dalmn hesaplayalm
medal_distribution_by_country = data.pivot_table(index='NOC', columns='Medal', aggfunc='size', fill_value=0)

plt.figure(figsize=(14, 8))
sns.heatmap(medal_distribution_by_country, annot=True, cmap='YlGnBu', fmt='d')
plt.title('lkelere Gre Madalya Trlerinin Dalm')
plt.xlabel('Madalya Tr')
plt.ylabel('lke Kodu')
plt.show()
```

```python
# Usain Leo Bolt'un performansn yllara gre inceleyelim
aftink_performance = data[data['Name'] == 'Usain St. Leo Bolt'].groupby('Year')['Medal'].count()

plt.figure(figsize=(12, 6))
aftink_performance.plot(kind='line', marker='o')
plt.title('Christine Jacoba Aaftinkin Yllara Gre Performans')
plt.xlabel('Yl')
plt.ylabel('Madalya Says')
plt.grid(True)
plt.show()
```

```python
# Spor dallarnn madalya kazanma oranlarn hesaplayalm
sport_medal_ratio = data.groupby('Sport')['Medal'].apply(lambda x: (x != 'NA').sum() / len(x))

plt.figure(figsize=(12, 8))
sport_medal_ratio.sort_values(ascending=False).plot(kind='bar')
plt.title('Spor Dallarnda Madalya Kazanma Oranlar')
plt.xlabel('Spor Dal')
plt.ylabel('Madalya Kazanma Oran')
plt.xticks(rotation=90)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Fizyolojik verileri seelim
features = data[['Age', 'Height', 'Weight']].dropna()

# Veriyi leklendirelim
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# K-means kmeleme
kmeans = KMeans(n_clusters=3, random_state=0)
data['Cluster'] = kmeans.fit_predict(scaled_features)

# Kmeleme sonularn grselletirelim
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Age', y='Height', hue='Cluster', data=data, palette='viridis')
plt.title('Sporcularn Ya ve Boya Gre Kmeleme')
plt.xlabel('Ya')
plt.ylabel('Boy')
plt.show()
```

```python
# Medalya kazanma orann hesaplayalm
data['Competition_Count'] = data.groupby('Name')['Name'].transform('count')
data['Medal_Efficiency'] = data['Medal_Code'] / data['Competition_Count']

# En verimli sporcular belirleyelim
top_athletes = data.groupby('Name')['Medal_Efficiency'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 8))
top_athletes.plot(kind='bar')
plt.title('En Verimli Sporcular')
plt.xlabel('Sporcu smi')
plt.ylabel('Madalya Kazanma Verimlilii')
plt.xticks(rotation=90)
plt.show()
```
